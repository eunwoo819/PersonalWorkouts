# 1. 네트워크 이론 기초


### 1.1. 네트워크란 무엇인가  

네트워크는 각 객체들이 상호 연결된 구조를 말한다. 네트워크는 노드 (Node, vertex)로 구성되고, 이 노드는 링크(Link, edge)로 연결된다.  
이때 노드들간의 관계를 링크라고 한다. 네트워크는 복수의 노드들이 연결된 구조라고 생각할 수 있다.
  
  

### 1.2. 네트워크 종류

**1) 방향성 / 가중치 유무에 따른 분류**  

네트워크는 링크의 방향성 유무와 가중치 유무에 따라 4가지 유형으로 구분 가능하다.
* 무방향(undirected) 네트워크 : 페이스북 친구 관계처럼 출발과 도착 지점 존재X
* 방향(directed) 네트워크 : 트위터 리트윗, 팔로잉 네트워크처럼 출발과 도착 지점 존재
* 이진(binary) 네트워크 : 관계의 유무만을 표현한 그래프
* 가중(valued) 네트워크 : 숫자로 가중치 표기 or 링크 선 굵기 조절의 방식으로 관계의 정도 나타낸 그래프

**2) 관계의 위상에 따른 분류**  
  
네트워크를 이루고 있는 관계 위상 형태에 따라  
Star, mesh, bus, ring, tree, Fully connected, Line 로 나눌 수 있다.  
  
    

**3) 네트워크 크기에 따른 분류**  
  
네트워크를 이루고 있는 **노드의 링크 규모 등 네트워크 크기**에 따라  
micro(소규모), meso(중규모), macro(대규모), super-macro(초대규모)로 나눌 수 있다.  

**4) 분석 대상에 따른 분류**  
  
* 전체 네트워크(complete network) : 네트워크를 구성하고 있는 모든 노드를 포함한 네트워크  
* 하위 네트워크(sub network) : 전체 네트워크의 일부분 (component, clique파당, cluster가 해당됨)

# 2.0. 소셜 네트워크 분석 지표

## 2.1. 네트워크 수준의 속성 분석
네트워크 수준에서는 네트워크의 크기, 밀도, 포괄성을 바탕으로 분석한다.   

* **네트워크 크기(network size)**   

: 네트워크를 구성하는 노드의 수  
: 방향성이 있는 경우 nx(n-1)로, 방향성이 없는 경우 nx(n-1)/2로 연결관계 수를 파악한다.  
  
* **네트워크 밀도(network density)**  
  
: 네트워크를 구성하고 있는 노드들 간의 연결된 정도  
: 네트워크 밀도는  
- 무방향 이진 네트워크 밀도 : k/{nx(n-2)/2}  
- 방향 이진 네트워크 밀도 : k/{nx(n-1)}  
로 구할 수 있다. 이때 k는 실제 연결된 링크의 수를, n은 노드의 수를 말한다.    

    
* **포괄성(inclusiveness)**  
: 네트워크 상의 총 노드 개수 중, 실제 연결 관계가 존재하는 노드의 비율을 뜻한다. 
: 포괄성은 (총 노드의 수 - 고립된 노드의 수 / 총 노드의 수) 로 구할 수 있다.

## 2.2. 노드 수준의 속성 분석
노드 수준에서는 연결정도, 연결거리, 직경, 평균연결거리를 바탕으로 분석한다.  
  
    
* **연결정도(degree)**  
: 해당 노드에 직접 연결되어 있는 노드들의 개수, 또는 링크의 개수  
: 고립된 노드의 경우는 연결정도가 0이 된다.  
: 연결정도가 높다 = 다른 노드들과 많은 관계를 갖는다 = 활동성과 영향력이 높다  
  
  
  * 출처 : http://www.textom.co.kr/home/sub/view.php?id=paper&no=30  
  위 네트워크는 럭셔리 호텔에 대한 네트워크 분석을 시각화한 자료이다.
  Haeundae(해운대), travel, seoul, jeju, best, busan trip이 럭셔리 호텔과 가장 자주 검색되었음을 확인 할 수 있다.  
  haeundae, travel, seoul 등 빈도가 높은 해당 노드들의 연결정도도 높다는 것을 알 수 있다.  
  이와 반대로 in-flight의 경우 다른 어떤 노드와도 연결되지 않았다.  
  해당 노드는 고립된 노드로 연결정도가 0이고, 네트워크 내에서 영향력이 없다는 것을 알 수 있다.
    
* **연결거리(distance)**  
: 노드간 관계(링크)의 강도  
: 두 노드간 연결 거리가 짧을수록 노드간의 연결이 높아진다.  
: 정보가 얼마나 신속하게 전달되는지 가늠하는 지표    
  
    
    
  연결거리도 같은 그래프로 확인하면, haeundae와 best, grand hotel, seoul, jeju, travel 노드는 가까이에 위치하지만, bus, morning, vacation, daegu, incheon 등 노드는 연결거리가 먼 것을 볼 수 있다.  
  
    
* **직경(diameter)**  
: 네트워크 상의 모든 연결거리 중에서 가장 긴 연결거리  
: 네트워크 상에서 가장 멀리 떨어져있는 두개 노드의 연결거리 = 네트워크 지름  
: 네트워크 직경이 작을수록 신속한 정보 전달이 가능해지고 효율성 높아짐  
    
     
      
   직경도 같은 그래프로 확인하면, work 노드와 airbusan 노드가 가장 긴 연결거리를 갖는다. 이 둘의 연결거리가 네트워크의 지름이고 해당 네트워크 직경이 작아야 신속한 정보 전달이 가능하고 효율성이 높아진다.
        
    
* **평균 연결거리(average distance)**  
: 네트워크 내 모든 노드 쌍의 최단 연결거리를 산출하여 평균값을 계산한 것  
: 네트워크 임의의 두 노드가 평균적으로 최소 몇 단계의 링크를 걸쳐 연결이 가능한지 보여준다.

## 2.3. 네트워크에 내재된 속성 분석

* **상호성(reciprocity)**  
: 방향 네트워크 상에서 2개의 노드 간 연결이 서로 쌍방향인 경우 **상호적 연결** 이라 하고, 단방향인 경우 **비상호적 연결** 이라고 한다.  
: 상호성 계산은 전체 네트워크 상에서 상호적 연결이 차지하고 있는 비율로 산출  
  
    
* **군집화 계수(clustering coefficient)**  
: 군집화의 정도를 나타내는 지표  
: 해당 노드의 이웃 노드들 간에 가능한 최대 관계의 수 중, 실제로 존재하는 관계의 수가 차지하는 비율  

# 3.0. 소셜 네트워크 분석 지표 - 중심성 분석
: 중심성(centrality)에는 **연결정도 중심성(degree centrality), 근접 중심성(closeness centrality), 매개 중심성(betweenness centrality), 아이겐 벡터 중심성(eigenvector centrality)** 가 있다.  
: 중심성 지표는 한 행위자가 전체 네트워크에서 중심에 위치하는 정도로 표현하는 지표이다.  
: 네트워크에서 중심적인 역할을 하는 구성원을 탐색하고, 그 구성원을 통해 효율적인 정보 전달과 의사전달을 가능하게 해 문제를 효과적으로 해결할 수 있다.  
  
    
## 3.1. 연결정도 중심성(degree centrality)  
: 한 노드가 얼마나 많은 다른 노드들과 연결관계를 맺고 있는지 측정  
: 노드에 연결된 링크의 수가 많다 = 노드의 연결정도 중심성 높다  
: 연결관계가 많아지면, 다른 노드에 대한 의존성을 상대적으로 낮춘다. 즉, 결과적으로 네트워크 상에서 더 큰 영향력을 미친다.    

    
* 출처 : 2016 대한건축학회 춘·추계 학술발표대회 논문
  
  해당 네트워크에서는 **BIM, 1인가구, 유지관리, 시뮬레이션, 모듈러건축, 타당성 분석**의 연결정도 중심성이 높은 것을 볼 수 있다.

## 3.2. 근접 중심성(closeness centrality)
: 한 노드가 얼마나 전체 네트워크의 중심에 근접해 있는지 측정  
: 해당 노드가 전체 네트워크 상에서 얼마나 중앙에 위치하고 있는지 측정  
: 노드가 중앙에 가깝게 위치할 수록, 자원과 정보를 전체 네트워크로 신속하게 확산 가능  
  
  
전체 네트워크의 중심에 있는 B가 가장 높은 근접 중심성을 가지며,  
B의 좌우에서 하부 네트워크를 연결해주는 노드 G, L이 두번째로 높은 근접 중심성 값을 갖는다.

## 3.3. 매개 중심성(betweenness centrality)
: 한 노드가 다른 노드들 간의 네트워크 관계 형성에 있어서 중개자 혹은 매개자 역할을 얼마나 수행하는 지 측정하는 지표  
: 노드의 매개 중심성이 높을 수록 네트워크 내에서 정보 및 자원의 흐름에 높은 통제력을 갖는다.  
  
    
  노드 B가 제일 높은 상대적 매개 중심성 값을 보이며,  
  노드 B의 좌우에서 각 하부 네트워크 길목에 위치하는 노드 A,G,L,C가 그 다음으로 높은 중심값을 보인다.

## 3.4 아이겐 벡터 중심성(eigenvector centrality)
: 연결된 다른 노드들의 중요도를 함께 반영한 것  
: 중심성이 높은 노드들과 많이 연결된 노드일수록, 해당 노드의 아이겐벡터 중심성이 커짐  
: 소수 영향력 높은 노드에 연결된 영향력 >>> 다수 일반 노드에 연결된 영향력  
: 친구가 많은 지 보다, 중요한 친구가 많은 지 보고자 할 때 사용하는 지표
